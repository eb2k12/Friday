```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
import category_encoders as ce
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from sklearn import tree
```

```python
#read in training data and drop rows with null values
test= pd.read_csv('tcd ml 2019-20 income prediction training (with labels).csv', delimiter = ',')
testa = test.dropna()
```

```python
#read in test data and drop rows with null values
preds = pd.read_csv('tcd ml 2019-20 income prediction test (without labels).csv', delimiter = ',')
preds_a = preds.fillna(method = 'ffill')
preds_a = preds_a[['Instance', 'Year of Record', 'Gender', 'Age', 'Country', 'Size of City', 'Profession', 'University Degree', 'Body Height [cm]', 'Income']]
```

```python
X = testa[['Instance', 'Year of Record', 'Gender', 'Age', 'Country', 'Size of City', 'Profession', 'University Degree','Body Height [cm]', 'Income in EUR']]
```

```python
#combine data sets for encoding
full_data = pd.concat([X, preds_a], sort=False)
```

```python
y = full_data['Income in EUR']
```

```python
#encode data
encoder = ce.TargetEncoder(cols=['Country', 'Profession', 'University Degree', 'Gender'])
encoder.fit(full_data, y)
full_data = encoder.transform(full_data)
```

```python
#split data back into training and test sets
features = full_data.reset_index()
for i, data in features.iterrows():
    if data['Instance'] == 111994:
        break
    index = i
print(index)
train_features = features.iloc[:index+1, :]
test_features = features.iloc[index+1:, :]
```

```python
labels = full_data['Income in EUR']
labels = labels.dropna()
labels = np.array(labels)
```

```python
train_features = train_features.drop(['Instance', 'Income in EUR', 'Income'], axis = 1)
```

```python
test_features = test_features[['Instance', 'Year of Record', 'Gender', 'Age', 'Country', 'Size of City', 'Profession', 'University Degree','Body Height [cm]']]
```

```python
#normalise
scaler = MinMaxScaler()
scaler.fit(train_features, labels)
train_features = scaler.transform(train_features)
```

```python
#split into training/validation test sets (local rmse)
X_test, X_val, y_test, y_val = train_test_split(train_features, labels, test_size = 0.25, random_state = 42)
```

```python
#local 
clf = CatBoostRegressor(n_estimators = 1000, random_state = 42)
clf.fit(X_test, y_test);
```

```python
val = clf.predict(X_val)
```

```python
errors = val - y_val
```

```python
#local rmse
np.sqrt(((errors) ** 2).mean())
```

```python
#normalise test data
scaler = MinMaxScaler()
test_features = scaler.fit_transform(test_features)
test_features = pd.DataFrame(test_features)
```

```python
#train model
clf = CatBoostRegressor(n_estimators = 1000, random_state = 42)
clf = clf.fit(train_features, labels)
```

```python
#predict
income_pred = clf.predict(test_features)
```

```python
#send results to csv
df = pd.read_csv('tcd ml 2019-20 income prediction submission file-Copy2.csv')
df['Income'] = income_pred
df.to_csv('tcd ml 2019-20 income prediction submission file-Copy2.csv', encoding='utf-8', index = False)
```
